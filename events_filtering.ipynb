# PLAYER_WITH_TEAM

```python
# Load the data from player_with_team.json
with open('player_with_team.json', 'r') as file:
    data = json.load(file)

# Extracting the data for the three DataFrames

# For player_with_team DataFrame
timestamps = [event["Timestamp"] for event in data]
entity1_ids = [event["Entity1"]["id"] for event in data]
entity2_ids = [event["Entity2"]["id"] for event in data]
values = [event["Value"] for event in data]
player_with_team_df = pd.DataFrame({
    'Timestamp': timestamps,
    'Entity1_id': entity1_ids,
    'Entity2_id': entity2_ids,
    'Value': values
})

# For player DataFrame
player_ids = [event["Entity1"]["id"] for event in data]
first_names = [event["Entity1"]["first_name"] for event in data]
last_names = [event["Entity1"]["last_name"] for event in data]
races = [event["Entity1"]["race"] for event in data]
classes = [event["Entity1"]["class"] for event in data]
guilds = [event["Entity1"]["guild"] for event in data]
last_logins = [event["Entity1"]["last_login"] for event in data]
player_df1 = pd.DataFrame({
    'id': player_ids,
    'first_name': first_names,
    'last_name': last_names,
    'race': races,
    'class': classes,
    'guild': guilds,
    'last_login': last_logins
}).drop_duplicates().reset_index(drop=True)

# For team DataFrame
team_ids = [event["Entity2"]["id"] for event in data]
team_names = [event["Entity2"]["team_name"] for event in data]
kingdoms = [event["Entity2"]["kingdom"] for event in data]
n_members = [event["Entity2"]["n_members"] for event in data]
team_df = pd.DataFrame({
    'id': team_ids,
    'team_name': team_names,
    'kingdom': kingdoms,
    'n_members': n_members
}).drop_duplicates().reset_index(drop=True)
```

# PLAYER_WITH_NPC

```python
# Load the data from player_with_team.json
with open('player_with_npc.json', 'r') as file:
    data = json.load(file)


# For player_with_npc DataFrame
timestamps = [event["Timestamp"] for event in data]
entity1_ids = [event["Entity1"]["id"] for event in data]
entity2_ids = [event["Entity2"]["id"] for event in data]
values = [event["Value"] for event in data]
Additional_Entity_Type = [event.get("Additional Entity Type", None) for event in data]
additional_entity_ids = [event.get("Additional Entity", {}).get('id', None) for event in data]
player_with_npc_df = pd.DataFrame({
    'Timestamp': timestamps,
    'Entity1_id': entity1_ids,
    'Entity2_id': entity2_ids,
    'Value': values,
    'Additional_Entity_Type': Additional_Entity_Type,
    'Additional_Entity_id': additional_entity_ids    
})

# For player DataFrame
player_ids = [event["Entity1"]["id"] for event in data]
first_names = [event["Entity1"]["first_name"] for event in data]
last_names = [event["Entity1"]["last_name"] for event in data]
races = [event["Entity1"]["race"] for event in data]
classes = [event["Entity1"]["class"] for event in data]
guilds = [event["Entity1"]["guild"] for event in data]
last_logins = [event["Entity1"]["last_login"] for event in data]
player_df2 = pd.DataFrame({
    'id': player_ids,
    'first_name': first_names,
    'last_name': last_names,
    'race': races,
    'class': classes,
    'guild': guilds,
    'last_login': last_logins
}).drop_duplicates().reset_index(drop=True)

# For npc DataFrame
npc_ids = [event["Entity2"]["id"] for event in data]
npc_types = [event["Entity2"]["npc_type"] for event in data]
first_names = [event["Entity2"]["first_name"] for event in data]
last_names = [event["Entity2"]["last_name"] for event in data]
locations = [event["Entity2"]["location"] for event in data]
npc_df1 = pd.DataFrame({
    'id': npc_ids,
    'npc_type': npc_types,
    'first_name': first_names,
    'last_name': last_names,
    'location': locations
}).drop_duplicates().reset_index(drop=True)

# For item DataFrame
item_ids = [event.get("Additional Entity", {}).get('id', None) for event in data if event.get("Additional Entity Type") == 'item']
item_names = [event.get("Additional Entity", {}).get('item_name', None) for event in data if event.get("Additional Entity Type") == 'item']
item_types = [event.get("Additional Entity", {}).get('item_type', None) for event in data if event.get("Additional Entity Type") == 'item']
item_df1 = pd.DataFrame({
    'id': item_ids,
    'item_name': item_names,
    'item_type': item_types
}).drop_duplicates().reset_index(drop=True)


# For dialogue DataFrame
dialogue_ids = [event.get("Additional Entity", {}).get('id', None) for event in data if event.get("Additional Entity Type") == 'dialogue']
content = [event.get("Additional Entity", {}).get('content', None) for event in data if event.get("Additional Entity Type") == 'dialogue']
choice_options = [event.get("Additional Entity", {}).get('choice_options', None) for event in data if event.get("Additional Entity Type") == 'dialogue']
emotions = [event.get("Additional Entity", {}).get('emotion', None) for event in data if event.get("Additional Entity Type") == 'dialogue']
dialogue_df1 = pd.DataFrame({
    'id': dialogue_ids,
    'content': content,
    'choice_options': choice_options,
    'emotion': emotions
}).drop_duplicates().reset_index(drop=True)
```
